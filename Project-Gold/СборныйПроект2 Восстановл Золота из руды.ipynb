### Проверим, что эффективность обогащения рассчитана правильно. Расчитаем МАЕ

#### *Введем константы для расчета RECOVERY по формуле*

```python
C = gold_train['rougher.output.concentrate_au'] # доля золота в концентрате после флотации
F = gold_train['rougher.input.feed_au'] # доля золота в сырье/концентрате до флотации
T = gold_train['rougher.output.tail_au'] # доля золота в отвальных хвостах после флотации
```

```python
calc_recovery = ((C*(F-T))/(F*(C-T)))*100
calculated = pd.Series(calc_recovery)
```

```python
recovery = gold_train['rougher.output.recovery'] # RECOVERY
```

```python
def mae(target, predictions):
    error = sum(abs(target-predictions))/len(target)
    return error

print(round(mae(recovery, calculated), 4))
```

### *Концентрация всех элементов после первого этапа (Флотация) - увеличивается. Затем, на первой стадии очистки, концентрация Золота и Свинца продолжает повышаться, а концентрация Серебра - снижается, и на последней стадии - растет только концентрация Золота. Это связано с тем, что процесс разделения смеси на фракции, в большинстве случаев, осуществляется методам Осаждения или методом Центрифугирования и зависят от молекулярной и физической массы элементов. Поэтому "поведение" графиков распределения концентрации Золота и Свинца схожи, так как массы этих элементов очень близки и они распологаются по соседству в таблице Менделеева. Т.е., например, при Центрифугировании будут отделяться фракции содержащие Золото и Свинец в приблизительно равных долях. В тоже время, Серебро имеет достаточное различие по массе с Золотом и Свинцом, поэтому "поведение" графиков концентрации Серебра отличается. Значит, скорее всего, концентрация Серебра в Отвальных хвостах будет расти. Также на гистограммах видны выбросы с нулевой концентрацией, которые потребуется удалить.*

#### *Два параметра имеют близкие по значению Средние и дисперсию. Если распределение значений близко к нормальному, то можно провести t-test для проверки гипотезы равенства двух параметров*

```python
fig, ax = plt.subplots(figsize=[12,7])

a_heights, a_bins = np.histogram(gold_train['rougher.input.feed_size'],  bins=30, density=True, range=(0,100))
b_heights, b_bins = np.histogram(gold_test['rougher.input.feed_size'], bins=a_bins, density=True)

width = (a_bins[1] - a_bins[0])/3

ax.bar(a_bins[:-1], a_heights, width=width, facecolor='blue', label='train')
ax.bar(b_bins[:-1]+width, b_heights, width=width, facecolor='red', label='test')
ax.legend()
ax.set_title('Распределение значений в Обучающей и тестовой выборках');
#sns.despine(ax=ax, offset=10)
```

### Выдвенем и проверим гипотезу, что средние значения обучающей и тестовой выборки равны

*Гипотеза H0 - средние выборок равны\
Гипотеза На - средние выборок не равны*

```python
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = 0.05   

results = st.ttest_ind(train, test)

# тест двухсторонний: p-value НЕ делим на 2
pvalue = results.pvalue

print('p-значение: ', pvalue)

if pvalue < alpha:
    print("Отвергаем нулевую гипотезу: средние значения в обучающей и тестовой выборке НЕ равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу: средние значения в обучающей и тестовой выборке РАВНЫ")
```

```python
def smape(target, predict):
    error = ((abs(target-predict))/(((abs(target))+(abs(predict)))/2)).mean()
    return error
```
